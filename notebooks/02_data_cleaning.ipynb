```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/raw/train.csv')

print(df.shape)
print(f"Missing values : {df.isnull().sum().sum()}")
```

```python
#remove outliers
print(f"Before: {df.shape}")
df = df[~((df['GrLivArea'] > 4000) & (df['SalePrice'] < 300000))]
print(f"After : {df.shape}")
```

```python
#drop columns with more than 50% missing values
cols_to_drop = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'MasVnrType']
df.drop(columns=cols_to_drop, inplace=True)
print(f"Shape after dropping : {df.shape}")
print(f"Missing values remaining : {df.isnull().sum().sum()}")
```

```python
#Handle FireplaceQu (47% missing) -> fill with 'None'
df['FireplaceQu'] = df['FireplaceQu'].fillna('None')
print(f"FireplaceQu missing: {df['FireplaceQu'].isnull().sum()}")
```

```python
#all 5 garage columns missing for same 81 houses

#categorical garage column -> fill with 'None'
garage_cat = ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']
for col in garage_cat:
    df[col] = df[col].fillna('None')

#numerical garage column -> fill with 0
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(0)

print(f"Garage missing values: {df[garage_cat].isnull().sum().sum()}")
```

```python
#Basement categorical columns -> fill with 'None'
# only fill actually missing basement columns
bsmt_cat = ['BsmtQual', 'BsmtCond', 
            'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']
df[bsmt_cat] = df[bsmt_cat].fillna('None')

print(f"Basement missing: {df[bsmt_cat].isnull().sum().sum()}")
```

```python
# fill LotFrontage with median of same neighborhood (group-wise imputation)
df['LotFrontage'] = df.groupby('Neighborhood')['LotFrontage'].transform(
    lambda x: x.fillna(x.median())
)

print(f"LotFrontage missing: {df['LotFrontage'].isnull().sum()}")
```

```python
df.isnull().sum()[df.isnull().sum()>0]
```

```python
# MasVnrArea: numerical, fill with 0 
df['MasVnrArea'] = df['MasVnrArea'].fillna(0)

# Electrical : categorical, fill with most frequent (mode)
df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])

print(f"Total missing: {df.isnull().sum().sum()}")
```

```python
#drop Id columnm as it's not useful for modeling
df.drop(columns=['Id'], inplace=True)
print(df.shape)
```

```python
df.to_csv('../data/processed/housing_cleaned.csv', index=False)
print("Cleaned data saved!")
print(df.shape)
```
